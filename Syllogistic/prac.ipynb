```python
def kfold_result(aug_method, num_aug):
    for i in range(5):
        k_fold_path = f"/root/syllogistic/syllogistic-generation/logs/KFOLD_VALIDATION_20epochs_t5_{aug_method}_num_augmentation_{num_aug}_filtering_true/log-KFOLD_VALIDATION_20epochs_t5_{aug_method}_num_augmentation_{num_aug}_filtering_true_kfold({i}).csv"
        log = pd.read_csv(k_fold_path)
        log_filtered = log[["epoch",'val_bleu', 'val_loss', 'val_rouge1','val_rouge2', 'val_rougeL']]
        
        if i == 0:
            total_df = log_filtered
        else:
            total_df = total_df + log_filtered
        
    total_df = total_df/5 # fold 수
    total_df["bleu_rouge1,2,l"] = (total_df["val_bleu"]+total_df["val_rouge1"]+total_df["val_rouge2"]+total_df["val_rougeL"])
    total_df["bleu_rougel"] = (total_df["val_bleu"]+total_df["val_rougeL"])
    # total_df.to_csv(f"/root/logs/Kfold_total/20epoch_{aug_method}_{num_aug}_filtering_true.csv")

    return total_df
```
