```python
def kfold_result(aug_method, num_aug):
    for i in range(5):
        # /root/syllogistic/syllogistic-generation/logs/KFOLD_VALIDATION_20epochs_t5_aug_mix_num_augmentation_1_filtering_true_prem_making_epoch_7
        k_fold_path = os.path.join("/root/syllogistic/syllogistic-generation/logs",f"KFOLD_VALIDATION_20epochs_t5_{aug_method}_num_augmentation_{num_aug}_filtering_true_prem_making_epoch_7/log-KFOLD_VALIDATION_20epochs_t5_{aug_method}_num_augmentation_{num_aug}_filtering_true_prem_making_epoch_7_kfold({i}).csv" )
        log = pd.read_csv(k_fold_path)
        if i == 0:
            log = log.iloc[5:]
            log = log.reset_index()
            
        log_filtered = log[["epoch",'val_bleu', 'val_loss', 'val_rouge1','val_rouge2', 'val_rougeL']]
        
        if i == 0:
            total_df = log_filtered
        else:
            total_df = total_df + log_filtered
        
    total_df = total_df/5 # fold ìˆ˜
    total_df["bleu_rouge1,2,l"] = (total_df["val_bleu"]+total_df["val_rouge1"]+total_df["val_rouge2"]+total_df["val_rougeL"])
    total_df["bleu_rougel"] = (total_df["val_bleu"]+total_df["val_rougeL"])
    # total_df.to_csv(f"/root/logs/Kfold_total/30epoch_{aug_method}_{num_aug}_filtering_{filtering}.csv")

    return total_df
```
